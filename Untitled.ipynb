```python
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression
import pandas as pd

# One-hot encode categorical features
x_train = pd.get_dummies(train_data.drop(['median_house_value'], axis=1))
y_train = train_data['median_house_value']

# Impute missing values
imputer = SimpleImputer(strategy="mean")
x_train_imputed = imputer.fit_transform(x_train)

# Create and train the Linear Regression model
reg = LinearRegression()
reg.fit(x_train_imputed, y_train)
```

```python
test_data = x_test.join(y_test)

# Log transformation of specific columns to reduce skewness
test_data['total_rooms'] = np.log(test_data['total_rooms'] + 1)
test_data['total_bedrooms'] = np.log(test_data['total_bedrooms'] + 1)
test_data['population'] = np.log(test_data['population'] + 1)
test_data['households'] = np.log(test_data['households'] + 1)

# One-hot encode the 'ocean_proximity' column and drop the original column
test_data = test_data.join(pd.get_dummies(test_data.ocean_proximity)).drop(['ocean_proximity'], axis=1)

# Create new ratio-based features
test_data['bedroom_ratio'] = test_data['total_bedrooms'] / test_data['total_rooms']
test_data['household_rooms'] = test_data['total_rooms'] / test_data['households']
```

```python
from sklearn.impute import SimpleImputer

# Impute missing values in x_test
imputer = SimpleImputer(strategy="mean")
x_test_imputed = imputer.fit_transform(x_test)

# Now you can calculate the score
score = reg.score(x_test_imputed, y_test)
print("Model score:", score)
```

```python
# Drop rows with NaN values
x_test_dropped = x_test.dropna()
y_test_dropped = y_test[x_test_dropped.index]  # Match y_test with the remaining x_test rows

# Calculate the score
score = reg.score(x_test_dropped, y_test_dropped)
print("Model score:", score)
```
